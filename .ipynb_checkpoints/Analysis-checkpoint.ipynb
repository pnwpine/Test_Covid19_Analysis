# Create Proof of Concept

```python
from ipyleaflet import Map, basemaps, basemap_to_tiles, Marker , CircleMarker
```

```python
center = ['31.8257','117.2264']
zoom = 7
m = Map(basemap=basemaps.CartoDB.Positron, center=center, zoom=zoom)
for lat,long,city in zip(covid19_confirmed['Lat'],covid19_confirmed['Long'],covid19_confirmed['Province/State']):
    
    mk_center = [lat,long]
    
    circle = Circle(stroke=False)
    circle.radius = 50000 #Will need to make some sort of scale.
    circle.color = "green"
    circle.fill_color = "green"
    circle.location = (mk_center)
    m.add_layer(circle)

    marker = Marker(location=mk_center, draggable=False, title=str(city), visible=False)
    m.add_layer(marker);

m
```

Ok! So POC is created and we can start to make the time series data visulization. 

Green -> Revovered
Red -> Dead
Yellow -> Infected
